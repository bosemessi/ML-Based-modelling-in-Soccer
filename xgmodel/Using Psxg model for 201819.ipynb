```python
import pickle
model = pickle.load(open('final_model.sav', 'rb'))
```

```python
# Calculate PSxG for 2018-19
# non_pen = pd.read_csv('postshots201819.csv', index_col=0)
non_pen = pd.read_csv('postshotsbar201819.csv', index_col=0)
non_pen = non_pen.drop(['y_loc','v1','v2','t','est_end_x','end_x_loc','distance_to_nearest_opponent',
                   'play_pattern_name','shot_technique_name','shot_first_time','under_pressure','direct_free_kick',
                    'distance', 'num_players_in_shot_cone'], axis=1).reset_index(drop=True)

features = non_pen.drop('goal', axis=1)
labels = non_pen['goal']
 
# Encode categorical features
cont_cols = ['duration', 'angle', 'shot_cone_angle', 'shot_cone',   
            'end_y_loc', 'end_z_loc', 'distance_to_nearest_opponent_na', 'cone_density', 'avg_vel', 'est_end_y']
 
cat_features = features.drop(cont_cols, axis=1)
cont_features = features[cont_cols]

def display_all(df):
    with pd.option_context("display.max_rows", 1000): 
        with pd.option_context("display.max_columns", 1000): 
            display(df.head(20).transpose())

# Use label encoding to convert categorical features to numeric
# https://stackoverflow.com/a/37038257
def label_encode(df):
    # Convert df to label encoded
    df_le = pd.DataFrame({col: df[col].astype('category').cat.codes for col in df}, index=df.index)
    # Save mappings as a dict
    mappings = {col: {n: cat for n, cat in enumerate(df[col].astype('category').cat.categories)} 
     for col in df}
    return df_le, mappings
 
cat_features_le, mappings = label_encode(cat_features)
 
# Merge with numeric features
features_le = cont_features.merge(cat_features_le, left_index=True, right_index=True)

X_new = features_le
y_new = labels

model.fit(X_new, y_new)

# Add xG and xA values using predict_proba
xg_df = X_new.copy()
xg_df['goal'] = y_new
xg_df['psxg'] = model.predict_proba(X_new)[:,1]
```
